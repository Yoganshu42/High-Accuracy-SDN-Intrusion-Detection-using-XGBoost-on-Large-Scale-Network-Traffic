```python
numerical_cols = df.select_dtypes(include=['int64','float64']).columns # checking for numerical and categorical columns
categorical_cols = df.select_dtypes(include=['object']).columns

print(f"Numerical Columns: {list(numerical_cols)}\n")
print(f"Categorical Columns: {list(categorical_cols)}")
```

```python
from sklearn.preprocessing import OneHotEncoder   #encoding the categorical columns
encoder = OneHotEncoder(sparse_output = False)
```

```python
encoded_class = encoder.fit_transform(df[['Class']])
```

```python
encoded_class
```

```python
class_names = encoder.categories_[0] #encoding the class column
encoded_df = pd.DataFrame(encoded_class, columns=class_names)
```

```python
# df = df.drop(columns = ['Class'])
df = pd.concat([df,encoded_df], axis = 1)
```

```python
print("Current Columns in DataFrame:\n", df.columns)
```
